```python
import pandas as pd
```

```python
path_direct="c:\workspace\sefactory\week1_pandas\kc_house_data.csv"

df=pd.read_csv(path_direct)
```

```python
all_columns=list(df.columns)
built_00= df.loc[(df['yr_built'] >= 2000), all_columns].copy()
built_00
```

```python
most_exp={}
for i in range(2000,2015):
    k=built_00[built_00.yr_built==i].price.max()
    most_exp[i]=k
most_exp
```

```python
mexp= pd.DataFrame.from_dict([most_exp]) #!!![most_exp]??
mexp
```

```python
most_bed={}
for i in range(2000,2015):
    k=built_00[built_00.yr_built==i].bedrooms.max()
    most_bed[i]=k
mbed=pd.DataFrame.from_dict([most_bed])
mbed
```

```python
most_bath={}
for i in range(2000,2015):
    k=built_00[built_00.yr_built==i].bathrooms.max()
    most_bath[i]=k
mbath=pd.DataFrame.from_dict([most_bath])
mbath
```

```python
most_flr={}
for i in range(2000,2015):
    k=built_00[built_00.yr_built==i].floors.max()
    most_flr[i]=k
mflr=pd.DataFrame.from_dict([most_flr])
mflr
```

```python
area_living = built_00.loc[(built_00.sqft_living > df.sqft_living.mean()), ['sqft_living','yr_built']].copy()
area_living
```

```python
area_lot= built_00.loc[(built_00.sqft_lot > df.sqft_lot.mean()), ['sqft_lot','yr_built']].copy()
area_lot
```

```python
built_append = pd.concat([mexp, mbed,mbath,mflr], axis=0, ignore_index=True)
built_append
```
